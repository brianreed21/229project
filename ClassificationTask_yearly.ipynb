```python
import numpy as np
import pandas as pd
import pickle

#####
# import relevant bits from keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers import Dropout
from keras import regularizers
from keras import optimizers

from sklearn.preprocessing import StandardScaler
import keras
from matplotlib import pyplot as plt
#####
# import relevant bits from sklearn

from sklearn.model_selection import train_test_split

from keras.models import load_model
```

# Prepare Data

## load

```python
data = pd.read_pickle('final_classification_yearly.pkl') # data_final = pd.read_pickle(folder + '/data_withInd.pkl')

y = data['Code_numeric']
X = data.drop(['firm','Description','Code',"Code_numeric"],axis = 1)


'''data = pd.read_pickle('final_classification.pkl') # data_final = pd.read_pickle(folder + '/data_withInd.pkl')

y = data['skdl1']
X = data.drop(['names','skdl1'],axis = 1)

'''
```

## convert Y to one-hot encoding

```python
ycat = keras.utils.to_categorical(y,num_classes = len(np.unique(y)))
```

```python
print(ycat.shape)
```

## split into train and test

```python
# split training into 
X_train, X_test, y_train, y_test = train_test_split(X, ycat, test_size=0.2)

# then split the test set into dev and test
X_test,X_dev, y_test,y_dev = train_test_split(X_test,y_test,test_size = 0.5)


scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_dev = scaler.fit_transform(X_dev)
X_test  = scaler.fit_transform(X_test)
```

## now define a very simple neural network
Make it a sequential model.

```python
# first, initialize the model
model = Sequential()
reg = 0.002
drop_rate = 0.5
model.add(Dense(1000,input_dim = X_train.shape[1]))
model.add(Dropout(drop_rate, noise_shape=None, seed=None))
model.add(Activation('relu'))

model.add(Dense(500,input_dim = X_train.shape[1]))
model.add(Dropout(drop_rate, noise_shape=None, seed=None))
model.add(Activation('relu'))

model.add(Dense(500,input_dim = X_train.shape[1]))
model.add(Dropout(drop_rate, noise_shape=None, seed=None))
model.add(Activation('relu'))

model.add(Dense(500,input_dim = X_train.shape[1]))
model.add(Dropout(drop_rate, noise_shape=None, seed=None))
model.add(Activation('relu'))

model.add(Dense(500,input_dim = X_train.shape[1]))
model.add(Dropout(drop_rate, noise_shape=None, seed=None))
model.add(Activation('relu'))

model.add(Dense(y_train.shape[1],activation = 'softmax'))
adam = optimizers.Adam(decay = 1e-6)
# For a multi-class classification problem
model.compile(optimizer=adam,loss='categorical_crossentropy',metrics=['accuracy'])

# experiment with different optimizersc
evals = model.fit(X_train,y_train,validation_data = (X_dev,y_dev),epochs = 500,batch_size = 512,verbose = 1)

model.save('adam_1e6decay_5do.h5')
with open('history_5', 'wb') as file:
        pickle.dump(evals.history, file)
```
